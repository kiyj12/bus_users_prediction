### 데이터 명세

- id : 해당 데이터의 고유한 ID (train, test와의 중복은 없음)

- date  : 날짜

- bus_route_id : 노선 id

- in_out  : 시내버스, 시외버스 구분 (categorical)

- station_code : 해당 승하차 정류소의 ID

- station_name : 해당 승하차 정류소의 이름

- latitude : 해당 버스 정류장의 위도 

- longitude : 해당 버스 정류장의 경도 

### *** 같은 위도경도여도 진행 방향에 따라 정류장이 다를수 있음.

- X-Y ride : 00:00부터 59:59까지 승차한 인원수

- X-Y takeoff : 00:00부터 59:59까지 하차한 인원수

- 18-20 ride : 18:00:00부터 19:59:59까지 승차한 인원 수 (train data에만 존재, 예측해야할 target)

```python
train['68_on']=train['6~7_ride']+train['7~8_ride'] # 6 ~ 8시 승차인원
train['810_on']=train['8~9_ride']+train['9~10_ride']
train['1012_on']=train['10~11_ride']+train['11~12_ride']

train['68_off']=train['6~7_takeoff']+train['7~8_takeoff'] # 6 ~ 8시 하차인원
train['810_off']=train['8~9_takeoff']+train['9~10_takeoff']
train['1012_off']=train['10~11_takeoff']+train['11~12_takeoff']

train.head()
```

### 시간대별 geton과 getoff 들의 상관성 확인

상관성이 생각보다 높게 나왔다.

```python
import seaborn as sns #시각화
import matplotlib.pyplot as plt # 시각화

train_onoff = train[['68_on', '810_on', '1012_on', '68_off', '810_off', '1012_off']]

cor = train_onoff.corr()

sns.set(style="white")
mask = np.zeros_like(cor, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f,ax = plt.subplots(figsize=(20,15))
cmap = sns.diverging_palette(200, 10, as_cmap=True)


# 히트맵 생성
sns.heatmap(cor, mask=mask, cmap = cmap, center = 0, square=True, linewidths=0.5, cbar_kws={'shrink':1}, annot = True)

plt.xticks(size = 15)
plt.yticks(size = 15, rotation= 0)
plt.title("arrive and leave correlation graph", size = 30)
```

```python
train3 = pd.DataFrame(columns={'6~7_ride','7~8_ride','8~9_ride','9~10_ride','10~11_ride','11~12_ride','18~20_ride'})
train3.insert(0, 'date', 0)


train3.head()
```

```python
datelist = train['date'].unique()

datelist = list(datelist)

datelist
```

```python
train3['date'] = datelist

train3.head()
```

```python
date_cnt = train2.groupby('date').size()

date_cnt = pd.DataFrame(date_cnt)
date_cnt = date_cnt.reset_index()


date_cnt.columns = ['date', 'day_count']
date_cnt.head()
```

```python
date_cnt['day_count'].sum()
```

```python
indexing = 0

for d in range(len(date_cnt)):
    cnt67 = 0
    cnt78 = 0
    cnt89 = 0
    cnt910 = 0
    cnt1011 = 0
    cnt1112 = 0
    cnt1820 = 0
    for i in range(date_cnt['day_count'][d]):
        cnt67 += train2['6~7_ride'][indexing + i]
        cnt78 += train2['7~8_ride'][indexing + i]
        cnt89 += train2['8~9_ride'][indexing + i]
        cnt910 += train2['9~10_ride'][indexing + i]
        cnt1011 += train2['10~11_ride'][indexing + i]
        cnt1112 += train2['11~12_ride'][indexing + i]
        cnt1820 += train2['18~20_ride'][indexing + i]
        
    indexing += date_cnt['day_count'][d]
    #print(date_cnt['day_count'][d])
    #print(indexing)
    train3.loc[d, '6~7_ride'] = cnt67
    train3.loc[d, '7~8_ride'] = cnt78
    train3.loc[d, '8~9_ride'] = cnt89
    train3.loc[d, '9~10_ride'] = cnt910
    train3.loc[d, '10~11_ride'] = cnt1011
    train3.loc[d, '11~12_ride'] = cnt1112
    train3.loc[d, '18~20_ride'] = cnt1820

    
train3.head()
```
